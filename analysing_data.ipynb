```python
import pandas as pd
df = pd.read_csv('Internship_Selection_Dataset.csv', sep=',', na_values=[''], quotechar='"')
df
```

```python
df['selected'] = df['selected'].replace({
    1:'Yes',
    0:'No',
})
```

```python
df.to_csv('Internship_Selection_Dataset.csv', index=False)
```

```python
df['github_score'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.bar(df['github_score'], df['selected'])

plt.ylabel('github_score')
plt.xlabel('Selected')
plt.show()
```
